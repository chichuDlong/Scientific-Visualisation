### 1D 线图——单图

#### 导入工具包，设置字体

```python
from math import *
import numpy as np
import matplotlib.pyplot as plt
# 添加下述指令，避免重复调用show()命令直接出图
%matplotlib inline
fontStyle = {'weight': 'normal', 'size': 20, }
languageUsed = 'Eng'
if languageUsed == "Eng":
    fontStyle = {'weight': 'normal', 'size': 20, }
else:
    fontStyle = {'family': 'SimHei', 'weight': 'normal', 'size': 20, }
```

#### 数据准备

```python
x=np.arange(0,2*pi,2*pi/100)
y=np.sin(x)
```

#### 绘制曲线

- <font size=4>设置画布大小</font>
- <font size=4>添加曲线标记(marker)、线型(line style)、颜色(color)；曲线标识(label)；线宽(linewidth,lw)；标记大小(markersize,ms)；标记间隔(markerevery)</font>
- <font size=4>添加坐标轴名称，设置字体样式、大小</font>
- <font size=4>设置tick大小、方向</font>
- <font size=4>设置图题</font>
- <font size=4>添加图例，设置图例位置、字体</font>
- <font size=4>设置x、y轴范围</font>
- <font size=4>设置图片网格、网格线型、颜色、线宽、透明度</font>
- <font size=4>添加文字，注释</font>

```python
plt.figure(figsize=(6.5,5))
plt.plot(x,y,"g*--",label="sin",lw=2,ms=10,markevery=6)
plt.xlabel("variable",fontsize=18)
plt.ylabel("function",fontsize=18)
plt.xticks(fontsize=18)
plt.yticks(fontsize=18)
plt.tick_params("both", which='major', length=5, width=1.5, colors='k', direction='in')  # "y", 'x', 'both'
plt.title("figure title",fontproperties=fontStyle, fontsize=20)
plt.legend(loc=3,prop=fontStyle)
plt.xlim(0,2*pi)
plt.ylim(-1,1)
plt.grid(linestyle="--",color="r",lw=0.5,alpha=0.5)
plt.text(3,0.5,"wen ben xin xi",fontsize=18)
plt.annotate("notation",xy=(3,0.1),xytext=(3.5,0.25),arrowprops=dict(facecolor="k",shrink=0.05),fontsize=18)
```

### 同一画布中多条曲线绘制

```python
x=np.arange(0,2*pi,2*pi/100)
y1=np.sin(x)
y2=np.cos(x)
plt.figure(figsize=(6.5,5))
plt.plot(x,y1,"g*--",label="sin",lw=2,ms=10,markevery=6)
plt.plot(x,y2,"b+-.",label="cos",lw=2,ms=10,markevery=6)
plt.xlabel("variable",fontsize=18)
plt.ylabel("function",fontsize=18)
plt.xticks(fontsize=18)
plt.yticks(fontsize=18)
plt.tick_params("both", which='major', length=5, width=1.5, colors='k', direction='in')  # "y", 'x', 'both'
plt.title("figure title",fontproperties=fontStyle, fontsize=20)
plt.legend(loc=3,prop=fontStyle)
plt.xlim(0,2*pi)
plt.ylim(-1,1)
plt.grid(linestyle="--",color="r",lw=0.5,alpha=0.5)
```

### 子图绘制 

```python
# 32*——要绘制的图形区域数(3*2=6)3行2列，从上到下，从左到右依次增加；*——为实际图形位置
fig=plt.figure(figsize=(6.5,5))
ax1=fig.add_subplot(321)
ax2=fig.add_subplot(326)
ax1.plot(x,y1,"g*--",label="sin",lw=2,ms=10,markevery=6)
ax2.plot(x,y2,"b+-.",label="cos",lw=2,ms=10,markevery=6)
fig.legend(loc=3,prop=fontStyle)# 多个子图公用一个图例，单个图例使用ax1.legend()
```
