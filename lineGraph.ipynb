### 1D 线图——单图

#### 导入工具包，设置字体

```python
from math import *
import numpy as np
import matplotlib.pyplot as plt
# 添加下述指令，避免重复调用show()命令直接出图
%matplotlib inline
fontStyle = {'weight': 'normal', 'size': 20, }
languageUsed = 'Eng'
if languageUsed == "Eng":
    fontStyle = {'weight': 'normal', 'size': 20, }
else:
    fontStyle = {'family': 'SimHei', 'weight': 'normal', 'size': 20, }
```

#### 绘制曲线
- 数据准备：x、y数据个数匹配
- 设置画布大小(英寸，默认$6.4\times4.8$)：`fig=plt.figure(figsize=(width,height))`
- 添加子图：`sub=fig.add_subplot(111)`，表示1行，1列，第1个图
>- 绘图：`sub.plot(x,y)`
>>- 设置曲线颜色(color)、线型(line style)、标记(marker)、曲线标签(label)；
>>- 线宽(linewidth,lw)；标记大小(markersize,ms)；标记间隔(markerevery)；
---
>- 添加坐标轴属性设置
>>- 1.坐标轴名称，字体样式、大小：`sub_xlabel("axisName",fontsize=Size)`<br>
>>- 2.坐标刻度范围：
>>>(a)、主刻度：`sub.set_xticks(np.arange(0,2*pi,2))`<br>
>>>(b)、小刻度：`sub.set_xticks(np.arange(0,2*pi,1),minor=True)`<br>
>>- 3.tick属性设置：
>>>(a)、对哪个轴进行设置：`axis="x" or "y" or "both"`<br>
>>>(b)、对主/副刻度设置：`which="major" or "minor" or "both"`<br>
>>>(c)、设置刻度长度：`length=size`<br>
>>>(d)、设置刻度宽度：`width=size`<br>
>>>(e)、设置刻度方向：`direction="in" or "out" or "inout"`<br>
>>>(f)、设置刻度标签距离轴的距离：`pad=size`<br>
>>>(g)、设置标签字体大小：`labelsize=size`<br>
>>>(h)、设置标签字体颜色：`labelcolor=color`<br>
>>>(i)、同时设置标签字体、刻度的颜色：`colors=color`<br>
>>>(j)、设置tick显示位置(`top`，`bottom`，`left`，`right`)，默认情况下仅有  `bottom`和`left`为`True`。<br>
>>>(k)、设置标签显示位置(`labeltop`，`labelbottom`，`labelleft`，`labelright`)，默认情况下仅有  `bottom`和`left`为`True`。<br>
>>>(l)、设置标签旋转角度：`labelrotation=angle`
---
- 添加子图图题：`sub.set_title("subTitleName",fontproperties=fontStyle,fontsize=size)`
- 添加整个图的标题：`fig.suptitle("titleName",fontproperties=fontStyle,fontsize=size)`
>- 添加图例
>>- 添加子图图例(loc=[posX,poY])，图例左下角在图中的位置：`sub.legend(loc=3,prop=fontStyle)`
>>- 添加全图图例：`fig.legend(loc=3,prop=fontStyle)`
---
>- 设置图片网格：
>>1.坐标轴设置：`axis="x" or "y" or "both"`<br>
>>2.指定刻度：`which="major" or "minor" or "both"`<br>
>>3.网格线型、颜色、线宽、透明度：`linestyle="--",color="r",lw=0.5,alpha=0.5`<br>
---
- 添加文字，包括文字内容、位置、大小、对齐方式、颜色、边框、填充色：<br>
`sub.text(0.5,0.75,"wen ben xin xi",fontsize=18,color="r",
horizontalalignment="center",verticalalignment="center",
transform=sub.transAxes,rotation=45,
bbox=dict(edgecolor="k",facecolor="None",alpha=0.3,
linestyle="--",linewidth=1.5))`
- 注释：`plt.annotate("notation",xy=(3,0.1),xytext=(3.5,0.25),arrowprops=dict(facecolor="k",shrink=0.05),fontsize=18)`<br>
通过命令`transform=sub.transAxes`将文本的位置改变对归一化参数
---
- 数据log显示：当数据非常大时，使用对数坐标显示式：<br>
`sub_set_yscale('log',base=10)或者sub_set_yscale('log',base=e)`
- 在实际绘图产生大量图时绘制保存完毕后关闭缓存中的图片：`plt.close(fig)`
---
>- 在给定图中绘制参考直线：<br>
>>- 水平线：`sub.axhline(y=posX,color=color,linestyle="--",linewidth=2)`
>>- 竖直线：`sub.axvline(y=posX,color=color,linestyle="-.")`
---

```python
x=np.arange(0,2*pi,2*pi/100)
y=np.sin(x)
fig=plt.figure(figsize=(6.5,5))
sub=fig.add_subplot(111)
sub.plot(x,y,"C1*--",label="sin",lw=2,ms=10,markevery=6)
sub.set_xlabel("variable",fontsize=18)
sub.set_ylabel("function",fontsize=18)
# 设置某一轴数据为log显示：sub_set_yscale('log',base=10)或者sub_set_yscale('log',base=e)
plt.xticks(fontsize=18)
plt.yticks(fontsize=18)
sub.set_xticks(np.arange(0,2*pi,0.5),minor=True)
plt.tick_params("both", which='major', length=5, width=1.5, colors='k', direction='in')  # "y", 'x', 'both'
plt.tick_params("both", which='minor', length=3, width=1.5, colors='r', direction='in')  # "y", 'x', 'both'
sub.set_title("figure title",fontproperties=fontStyle, fontsize=20)
fig.suptitle("Total Title",fontsize=20)
sub.axhline(y=0,color="k",linestyle="--",linewidth=2)
sub.axvline(x=2,color="g",linestyle="-.",linewidth=2)
plt.legend(loc=3,prop=fontStyle)
plt.xlim(0,2*pi)
plt.ylim(-1,1)
plt.grid(linestyle="--",color="r",lw=0.5,alpha=0.5)
sub.text(0.5,0.75,"wen ben xin xi",fontsize=18,color="r",
horizontalalignment="center",verticalalignment="center",transform=sub.transAxes,
rotation=45,bbox=dict(edgecolor="k",facecolor="None",alpha=0.3,linestyle="--",
linewidth=1.5))
plt.annotate("notation",xy=(3,0.1),xytext=(3.5,0.25),arrowprops=dict(facecolor="k",shrink=0.05),fontsize=18)
plt.tight_layout()
```

#### 双y轴曲线
- 首先绘制第一个曲线，然后设置好相关的标题、图名、坐标轴等等参数
- 通过设置共享x轴实现双y轴：`sub2=sub.twinx()`
- 使用相同的操作对sub2进行相关设置，
- 设置曲线对应y轴、y轴字体、刻度颜色、大小：<br>
`sub2.spines["left"].set_color("red")`<br>
`sub2.spines["right"].set_color("b")`<br>

```python
x=np.arange(0,2*pi,2*pi/100)
y1=np.sin(x)
y2=x**2
fig=plt.figure(figsize=(6.5,5))
sub=fig.add_subplot(111)
sub.plot(x,y1,"g*--",label="sin",lw=2,ms=10,markevery=6)
sub.set_xlabel("x",fontsize=18)
sub.set_ylabel("y1",fontsize=18,color="g")
sub.set_xticks(np.arange(0,2*pi,2))# 主刻度范围设置
sub.set_xticks(np.arange(0,2*pi,1),minor=True) # 小刻度设置
sub.tick_params(axis="y",length=8,width=2.3,direction="in",color="g",
labelsize=18,labelcolor="g")
sub.tick_params(axis="x",which="major",length=8,width=2.3,direction="in",color="r")
sub.set_title("figure title",fontproperties=fontStyle, fontsize=20)
sub.legend(loc=3,fontsize=18)
sub.grid(axis="both",linestyle="--",color="r",lw=0.5,alpha=0.5) 
sub2=sub.twinx()
sub2.plot(x,y2,"b+-.",label="$y=x^2$",lw=2,ms=10,markevery=6)
sub2.tick_params(axis="y",length=8,width=2.3,direction="in",color="b",
labelsize=18,labelcolor="b")
sub2.legend(loc=1,fontsize=18)
sub2.spines["left"].set_color("red")
sub2.spines["right"].set_color("b")
fig.legend()
plt.tight_layout()
```

#### 添加曲线辅助阴影区域
> 可以通过添加阴影区域，为阴影区域添加图例的方式对数据中的特殊部分进行强调分析。
>- 添加竖直阴影区域：
>-`axvspan(x0,x1,y0,y1,color=color,alpha=0.2,label="name")`<br>
>- x0、x1为实际横坐标范围，y0、y1为归一化的值
>- 添加水平阴影区域：
>-`axhspan(x0,x1,y0,y1,color=color,alpha=0.2,label="name")`<br>
>- x0、x1为实际纵坐标范围，y0、y1为归一化的值
>- 填充介于给定的区域——填充介于y0(x)~y1(x)之间的区域，x、y0、y1可以通过函数实现：
>- `fill_between(x,y0,y1)`或者
>- `fill_between(funcX,funcY0,funcY1)`

```python
x=np.linspace(0,2,301)
y=np.cos(2*pi*x)
fig=plt.figure()
sub=fig.add_subplot(111)
sub.plot(x,y,label="y=cos(x)")
sub.set_xlabel("x")
sub.set_ylabel("y")
sub.axvspan(0.25,0.75,0,0.5,color="y",alpha=0.5,label="$y\leq 0$")
sub.axhspan(0.5,0.75,0.25,0.75,color="r",alpha=0.5,label="$y\geq 0$\n$0.5<x<0.75$")
sub.fill_between([1.5,1.75],[-1,-1],[0,0],color="g",alpha=0.3)
sub.legend()
```

### 同一画布中多条曲线绘制
> 多个图形，图例选择性绘制：<br>
`line1,=sub.plot(x,y1,"g*--",label="sin",lw=2,ms=10,markevery=6)`<br>
`line2,=sub.plot(x,y2,"b+-.",label="cos",lw=2,ms=10,markevery=6)`<br>
`sub.legend(handles=[line1],labels=["xian tiao 1"])`<br>
为了能够显示多个图例防止覆盖，采用如下操作，将图例固定到图片上成为静态图片。
`leg1=sub.legend(handles=[line1],labels=["xian tiao 1"])`<br>
`sub.add_artist(leg1)`
`sub.legend(handles=[line2])`

```python
x=np.arange(0,2*pi,2*pi/100)
y1=np.sin(x)
y2=np.cos(x)
fig=plt.figure(figsize=(6.5,5))
sub=fig.add_subplot(111)
line1,=sub.plot(x,y1,"g*--",label="sin",lw=2,ms=10,markevery=6)
line2,=sub.plot(x,y2,"b+-.",label="cos",lw=2,ms=10,markevery=6)
sub.set_xlabel("variable",fontsize=18)
sub.set_ylabel("function",fontsize=18)
sub.set_xticks(np.arange(0,2*pi,2))# 主刻度范围设置
sub.set_xticks(np.arange(0,2*pi,1),minor=True) # 小刻度设置
sub.tick_params(axis="x",which="major",length=8,width=2.3,direction="in",color="r",
labelsize=18,labelcolor="C2")
plt.yticks(np.arange(-1,1,0.5),fontsize=18)
sub.set_title("figure title",fontproperties=fontStyle, fontsize=20)
leg1=sub.legend(loc=3,handles=[line1],prop=fontStyle)
sub.add_artist(leg1)
sub.legend(loc=1,handles=[line2],fontsize=18)
plt.xlim(0,2*pi)
plt.ylim(-1,1)
plt.grid(axis="x",linestyle="--",color="r",lw=0.5,alpha=0.5) 
```

### 子图绘制 

```python
# 32*——要绘制的图形区域数(3*2=6)3行2列，从上到下，从左到右依次增加；*——为实际图形位置
x=np.arange(0,2*pi,2*pi/100)
y1=np.sin(x)
y2=np.cos(x)
fig=plt.figure(figsize=(6.5,5))
sub1=fig.add_subplot(321)
sub2=fig.add_subplot(326)
sub1.plot(x,y1,"g*--",label="sin",lw=2,ms=10,markevery=6)
sub2.plot(x,y2,"b+-.",label="cos",lw=2,ms=10,markevery=6)
sub1.set_title("sub1",fontsize=15)
sub2.set_title("sub2",fontsize=15)
fig.suptitle("zong biao ti",fontsize=20)
fig.legend(loc=3,prop=fontStyle)# 多个子图公用一个图例，单个图例使用sub1.legend()
```
