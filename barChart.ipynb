### 柱形图

#### 基本柱形图

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5,6]
y=[34,67,58,87,23,49]
plt.bar(x,y,label="basic",color=["r","g","b","y","c","m"],ec="k",lw=1.5,ls="--",hatch="//")
plt.title("Bar chart",fontsize=18)
plt.xlabel("year",fontsize=18)
plt.ylabel("Income",fontsize=18)
plt.xticks(fontsize=18)
plt.yticks(fontsize=18)
plt.tick_params("both", which='major', length=8, width=1.5, colors='k', direction='in')
plt.legend(prop={"weight":"normal","size":18})
plt.show()
```

#### 堆叠柱形图

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5,6]
y1=[34,67,58,87,23,49]
y2=[18,20,31,49,64,57]
plt.bar(x,y1,label="A",bottom=y2,color="b",ec="k",lw=1.5,ls="--",hatch="//")
plt.bar(x,y2,label="B",color="y",ec="r",lw=1.5,ls="-.",hatch="\\")
#plt.bar(x,y1,label="A",color="b",ec="k",lw=1.5,ls="--",hatch="//")
#plt.bar(x,y2,label="B",bottom=y1,color="y",ec="r",lw=1.5,ls="-.",hatch="\\")
plt.title("Bar chart",fontsize=18)
plt.xlabel("year",fontsize=18)
plt.ylabel("Income",fontsize=18)
plt.xticks(fontsize=18)
plt.yticks(fontsize=18)
plt.tick_params("both", which='major', length=8, width=1.5, colors='k', direction='in')
plt.legend(prop={"weight":"normal","size":18})
plt.show()
```

### 并列柱形图

```python
import matplotlib.pyplot as plt
import numpy as np
barWidth=0.35
x1=np.arange(1,7)
x2=x1+barWidth
y1=[34,67,58,87,23,49]
y2=[18,20,31,49,64,57]
tickLabel=["A","B","C","D","E","F"]
plt.bar(x1,y1,label="A",color="b",ec="k",lw=1.5,ls="--",hatch="//",width=barWidth)
plt.bar(x2,y2,label="B",color="y",ec="r",lw=1.5,ls="-.",hatch="\\",width=barWidth)
plt.title("Bar chart",fontsize=18)
plt.xlabel("year",fontsize=18)
plt.ylabel("Income",fontsize=18)
# 设置x轴标签位置，及标签名称，标签旋转用来显示多字符
plt.xticks(x1+barWidth/2,tickLabel,rotation=90,fontsize=18)
plt.yticks(fontsize=18)
plt.tick_params("both", which='major', length=8, width=1.5, colors='k', direction='in')
plt.legend(prop={"weight":"normal","size":18})
plt.show()
```

#### 中文乱码问题

```python
import matplotlib.pyplot as plt
import matplotlib as mpl
#mpl.rcParams["font.sans-serif"]=["SimHei"]
mpl.rcParams["font.sans-serif"]=["STXinwei"]
mpl.rcParams["axes.unicode_minus"]=False
x=[1,2,3,4,5,6]
y=[34,67,58,87,23,49]
plt.bar(x,y,label="基本",color=["r","g","b","y","c","m"],ec="k",lw=1.5,ls="--",hatch="//")
plt.title("柱形图",fontsize=18)
plt.xlabel("年份",fontsize=18)
plt.ylabel("收入",fontsize=18)
plt.xticks(fontsize=18)
plt.yticks(fontsize=18)
plt.tick_params("both", which='major', length=8, width=1.5, colors='k', direction='in')
plt.legend(prop={"weight":"normal","size":18})
plt.show()
```
