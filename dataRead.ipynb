### Python数据读取——numpy

#### 1、文本中数据为2列纯数字：`np.loadtxt("dataPath")`

```python
import matplotlib.pyplot as plt
import numpy as np
data=np.loadtxt("./data.txt")
print(data.shape)
print(data)
```

##### 将上述读取的数据直接保存至x、y列表中

```python
x=data[:,0]
y=data[:,1]
plt.plot(x,y)
plt.show()
```

##### 2、数据前带有标签提示：`np.loadtxt("dataPath",skiprows=num)`

```python
data1=np.loadtxt("./data1.txt",skiprows=1)
print(data1)
```

##### 3、通过给定分割符，读取不同数据格式(txt,csv)的数据文件：`np.loadtxt("dataPath",skiprows=num,delimiter=",")`

```python
data2=np.loadtxt("./data2.csv",skiprows=1,delimiter=",")
print(data2)
```

#### 1、读取Excel表格中的数据：`pd.read_excel("dataPath",sheet_name="工作表名")`，也可以是0、1、2等整数，表示第1个工作表···
