### 饼图
#### 饼图中各个参数说明
- x：饼图中每一块的比例，通常为一个数组。如果sum(x)>1则使用归一化比例，即每一块除以sum(x)，如果sum(x)<1，则按照实际比例，此时有一部分为空白；
- explode：指定饼图中每块离开中心的距离，通常为一个数组，默认为0；
- labels：为饼图添加标签说明，类似于图例说明，标签列表；
- colors：指定饼图的填充色，颜色列表
- autopct：设置饼图的百分比样式，可以用format格式化指定饼图内百分比数字显示宽度与小数位数；
- radius：设置饼图的半径；
- labeldistance：每块旁边的文本标签的位置离饼的中心点的距离；
- shadow：是否有阴影，默认为False

```python
import matplotlib.pyplot as plt
x=[2,3,4,5]
labels=["A","B","C","D"]
colors=["r","g","b","y"]
explode=[0,0,0,0.1]
fontStyle={"color":"c","fontsize":18,"fontfamily":"SimHei"}
wedgeprops={"edgecolor":"r","linestyle":"-","alpha":0.9}
plt.figure(figsize=(7,7))
plt.title("pie Chart",color="red",size=18)
plt.pie(x,labels=labels,labeldistance=1.1,explode=explode,colors=colors,
autopct="%.2f%%",radius=0.8,textprops=fontStyle,wedgeprops=wedgeprops)
plt.legend(loc="upper right",title="pie learning",ncol=2,prop={"size":15})
plt.tight_layout()
plt.show()
```

#### 修改每一个饼图返回值的字体
plt.pie()返回3个参数，分别为：wedge(饼图)，texts(标签)，autotexts(百分比)

```python
import matplotlib.pyplot as plt
x=[2,3,4,5]
labels=["A","B","C","D"]
colors=["r","g","b","y"]
explode=[0,0,0,0.1]
fontStyle={"color":"c","fontsize":18,"fontfamily":"Fangsong"}
wedgeprops={"edgecolor":"r","linestyle":"-","alpha":0.9}
plt.figure(figsize=(7,7))
plt.title("pie Chart",color="red",size=18)
wedges,texts,autotext=plt.pie(x,labels=labels,labeldistance=1.1,explode=explode,colors=colors,
autopct="%.2f%%",radius=0.8,textprops=fontStyle,wedgeprops=wedgeprops)
for i in range(len(texts)):
    texts[i].set_color("g")
    texts[i].set_size(30)
    texts[i].set_weight("bold")
for i in range(len(autotext)):
    autotext[i].set_color("w")
    autotext[i].set_size(25)
plt.legend(loc="upper right",title="pie learning",ncol=2,prop={"size":15})
plt.tight_layout()
plt.show()
```
